# tse_snapshot_day.ipynb
## TSE全銘柄スナップショット -- 指定1日のQVTスコア一括生成

```
build_tse_master（銘柄マスタ）
    down  市場フィルタ・除外設定
TSE対象銘柄リスト
    down  fetch_jquants_cache.py（オプション）
J-Quantsファンダメンタルキャッシュ
    down  backtest_exportの計算ロジック流用
QVTスコア全銘柄分
    down
output/snapshot_YYYYMMDD.csv
```

````python
# ===================================================
# Cell 1: Config (edit here only)
# ===================================================

# Snapshot date (YYYY-MM-DD)
TARGET_DATE = "2024-03-07"   # <- change this

# Market scope: all / prime / standard / growth
MARKET_SCOPE = "all"

# Exclusion flags
EXCLUDE_FINANCIAL = False   # exclude banks, insurance etc.
EXCLUDE_ETF_REIT  = True    # exclude ETF / REIT

# J-Quants credentials (leave empty to use yfinance fallback)
JQUANTS_EMAIL    = "the3stigmataofpalmereldritch@gmail.com"   # e.g. "your@email.com"
JQUANTS_PASSWORD = "469Maonethousand"   # e.g. "your_password"

# Force J-Quants cache rebuild
JQUANTS_REFRESH_CACHE = False

# yfinance sleep interval (sec)
SLEEP_SEC = 0.5

# Rebuild TSE master from scratch
REFRESH_MASTER = False

# Output directory
OUTPUT_DIR = "output"

# Parallel processing
USE_PARALLEL = False
MAX_WORKERS  = 4

# ---- validation ----
import re
from datetime import datetime
assert re.match(r'\d{4}-\d{2}-\d{2}', TARGET_DATE), 'TARGET_DATE must be YYYY-MM-DD'
TARGET_DT = datetime.strptime(TARGET_DATE, '%Y-%m-%d')

print(f'Snapshot date: {TARGET_DATE}')
print(f'Market scope : {MARKET_SCOPE}')
print(f'Excl. finl.  : {EXCLUDE_FINANCIAL}')
print(f'Excl. ETF    : {EXCLUDE_ETF_REIT}')
print(f'Output CSV   : {OUTPUT_DIR}/snapshot_{TARGET_DATE.replace("-","")}.csv')
````

````python
# ===================================================
# Cell 2: Path setup & imports
# ===================================================

import os, sys, time, warnings, traceback
from datetime import datetime, timedelta
from pathlib import Path
from typing import Optional, Dict, Any, List

import pandas as pd
import numpy as np
import yfinance as yf

warnings.filterwarnings('ignore')

NOTEBOOK_DIR = Path().resolve()
for _p in [str(NOTEBOOK_DIR), str(NOTEBOOK_DIR / 'app')]:
    if _p not in sys.path:
        sys.path.insert(0, _p)

# checkSIGNALs modules
from modules.q_logic    import score_quality
from modules.valuation  import score_valuation
from modules.t_logic    import compute_t_metrics
from modules.indicators import (
    calc_moving_averages, calc_bollinger_bands, calc_rsi, calc_slope
)

# TSE master builder
from build_tse_master import build_tse_master

Path(OUTPUT_DIR).mkdir(exist_ok=True)
print('imports OK')
````

````python
# ===================================================
# Cell 3: Build TSE ticker universe
#   Reuses build_tse_master to get the full ticker list.
#   REFRESH_MASTER=False -> reuse existing tse_master_latest.csv
# ===================================================

_master_local = Path(OUTPUT_DIR) / 'tse_master_latest.csv'

if not REFRESH_MASTER and _master_local.exists():
    print(f'Reusing existing master: {_master_local}')
    master_df = pd.read_csv(_master_local, encoding='utf-8-sig')
else:
    print('Running build_tse_master...')
    result = build_tse_master(
        market_scope=MARKET_SCOPE,
        refresh_pool=REFRESH_MASTER,
        sleep_sec=0.5,
        use_yfinance=True,
        save_csv=True,
        output_dir=OUTPUT_DIR,
    )
    master_df = result['master_df']
    print(result['summary'])

print(f'Master rows: {len(master_df)}')

# -- market filter --
if MARKET_SCOPE != 'all' and 'market' in master_df.columns:
    scope_keywords = {
        'prime':    ['Prime Market', 'prime', 'PRIME'],
        'standard': ['Standard Market', 'standard', 'STANDARD'],
        'growth':   ['Growth Market', 'growth', 'GROWTH'],
    }
    kws = scope_keywords.get(MARKET_SCOPE, [])
    if kws:
        mask = master_df['market'].str.contains('|'.join(kws), na=False, case=False)
        master_df = master_df[mask].copy()
        print(f'After market filter ({MARKET_SCOPE}): {len(master_df)}')

# -- exclusion filters --
FINANCIAL_CATS = {'bank','insurance','leasing','securities','other_financial'}
if EXCLUDE_FINANCIAL and 'internal_category' in master_df.columns:
    before = len(master_df)
    master_df = master_df[~master_df['internal_category'].isin(FINANCIAL_CATS)].copy()
    print(f'After excl. financials: {before} -> {len(master_df)}')

if EXCLUDE_ETF_REIT and 'quote_type' in master_df.columns:
    before = len(master_df)
    master_df = master_df[
        ~master_df['quote_type'].str.lower().isin(['etf','reit','fund'])
    ].copy()
    print(f'After excl. ETF/REIT: {before} -> {len(master_df)}')

TICKERS: List[str] = [
    t for t in master_df['ticker'].dropna().unique().tolist()
    if t and str(t).strip()
]

print(f'\nTarget tickers: {len(TICKERS)}')
print(f'First 10: {TICKERS[:10]}')
print(f'Last  10: {TICKERS[-10:]}')
````

````python
# ===================================================
# Cell 4: J-Quants fundamentals cache
#   If credentials are empty -> falls back to yfinance
# ===================================================

import requests, json as _json

CACHE_PATH = Path(OUTPUT_DIR) / 'fundamentals_cache.csv'

if not JQUANTS_EMAIL or not JQUANTS_PASSWORD:
    print('No J-Quants credentials -> will use yfinance fallback')
    JQUANTS_REFRESH_TOKEN = ''

elif not JQUANTS_REFRESH_CACHE and CACHE_PATH.exists():
    print(f'Cache exists, skip rebuild: {CACHE_PATH}')
    print('Set JQUANTS_REFRESH_CACHE=True to force rebuild.')
    JQUANTS_REFRESH_TOKEN = '(cached)'

else:
    print('Fetching J-Quants refresh token...')
    _resp = requests.post(
        'https://api.jquants.com/v1/token/auth_user',
        data=_json.dumps({'mailaddress': JQUANTS_EMAIL, 'password': JQUANTS_PASSWORD}),
        headers={'Content-Type': 'application/json'},
    )
    if _resp.status_code != 200:
        raise ValueError(f'Auth failed: {_resp.status_code} {_resp.text}')
    JQUANTS_REFRESH_TOKEN = _resp.json().get('refreshToken', '')
    if not JQUANTS_REFRESH_TOKEN:
        raise ValueError(f'No refreshToken: {_resp.json()}')
    print('Refresh token OK')

    # Pass TICKERS to fetch_jquants_cache.py via module-level vars
    # The script reads TICKERS / JQUANTS_EMAIL / JQUANTS_PASSWORD / OUTPUT_DIR
    # from the calling namespace when invoked via %run
    print(f'Building fundamentals cache for {len(TICKERS)} tickers...')
    %run fetch_jquants_cache.py

print('Fundamentals cache step done.')
````

````python
# ===================================================
# Cell 5: Utility functions (ported from backtest_export.ipynb)
# ===================================================

NA = '-'  # missing value sentinel


def _safe(x, digits=2):
    try:
        if x is None or (isinstance(x, float) and np.isnan(x)):
            return NA
        return round(float(x), digits)
    except Exception:
        return NA


def get_close_col(df: pd.DataFrame) -> Optional[str]:
    for c in df.columns:
        if str(c).lower().strip() == 'close':
            return c
    for c in df.columns:
        if 'close' in str(c).lower():
            return c
    return None


def get_quarter_key(date_str: str) -> str:
    dt = datetime.strptime(date_str, '%Y-%m-%d')
    q  = (dt.month - 1) // 3 + 1
    return f'{dt.year}-Q{q}'


# -- Load J-Quants cache --
_fund_df: Optional[pd.DataFrame] = None
if CACHE_PATH.exists():
    _fund_df = pd.read_csv(CACHE_PATH, encoding='utf-8-sig', dtype=str)
    _num_cols = ['roe','roa','operating_margin','equity_ratio',
                 'de_ratio','interest_coverage','eps','bps','ev_ebitda']
    for col in _num_cols:
        if col in _fund_df.columns:
            _fund_df[col] = pd.to_numeric(_fund_df[col], errors='coerce')
    DATA_SOURCE = 'J-Quants (historical)'
    print(f'Cache loaded: {len(_fund_df)} rows / {_fund_df["ticker"].nunique()} tickers')
else:
    DATA_SOURCE = 'yfinance (current values only)'
    print(f'No cache found -> {DATA_SOURCE}')

print(f'Data source: {DATA_SOURCE}')


def _lookup_fund(ticker: str, quarter_key: str) -> Optional[Dict[str, Any]]:
    if _fund_df is None:
        return None
    df = _fund_df
    hit = df[(df['ticker'] == ticker) & (df['quarter_key'] == quarter_key)]
    if hit.empty:
        cands = df[(df['ticker'] == ticker) & (df['quarter_key'] <= quarter_key)]
        if cands.empty:
            return None
        hit = cands.sort_values('quarter_key', ascending=False).head(1)
    else:
        hit = hit.sort_values('disclosed_date', ascending=False).head(1)
    row = hit.iloc[0]
    keys = ['roe','roa','operating_margin','equity_ratio',
            'de_ratio','interest_coverage','eps','bps','ev_ebitda']
    return {k: (None if pd.isna(row[k]) else row[k]) for k in keys}


def _get_fundamentals_yf(ticker: str) -> Dict[str, Any]:
    result = {k: None for k in [
        'company_name','eps','bps','roe','roa','equity_ratio',
        'operating_margin','de_ratio','interest_coverage','ev_ebitda',
        'industry','sector',
    ]}
    try:
        info = yf.Ticker(ticker).info or {}
        def _f(key, scale=1.0):
            v = info.get(key)
            try: return float(v) * scale if v not in (None, '', 'None') else None
            except: return None
        result['company_name']     = info.get('longName') or info.get('shortName') or ticker
        result['eps']              = _f('trailingEps')
        result['bps']              = _f('bookValue')
        result['roe']              = _f('returnOnEquity',   100.0)
        result['roa']              = _f('returnOnAssets',   100.0)
        result['operating_margin'] = _f('operatingMargins', 100.0)
        result['de_ratio']         = (_f('debtToEquity', 1/100.0)
                                      if info.get('debtToEquity') else None)
        result['ev_ebitda']        = _f('enterpriseToEbitda')
        result['industry']         = info.get('industry', '')
        result['sector']           = info.get('sector',   '')
    except Exception:
        pass
    return result


def fetch_fund_for_date(ticker: str, date_str: str) -> Dict[str, Any]:
    qk     = get_quarter_key(date_str)
    result: Dict[str, Any] = {'_fetched_quarter': qk}

    cached = _lookup_fund(ticker, qk)
    if cached is not None:
        result.update(cached)
        result['_source'] = 'jquants_cache'
        try:
            info = yf.Ticker(ticker).info or {}
            result['company_name'] = info.get('longName') or info.get('shortName') or ticker
            result['industry']     = info.get('industry', '')
            result['sector']       = info.get('sector',   '')
        except Exception:
            result.setdefault('company_name', ticker)
            result.setdefault('industry', '')
            result.setdefault('sector',   '')
        return result

    yf_data = _get_fundamentals_yf(ticker)
    yf_data['_fetched_quarter'] = qk
    yf_data['_source'] = 'yfinance'
    return yf_data


def fetch_ohlcv_for_date(ticker: str, date_str: str,
                          lookback_days: int = 400) -> Optional[pd.DataFrame]:
    end_dt   = datetime.strptime(date_str, '%Y-%m-%d') + timedelta(days=1)
    start_dt = end_dt - timedelta(days=lookback_days)
    for _ in range(2):
        try:
            df = yf.download(
                ticker,
                start=start_dt.strftime('%Y-%m-%d'),
                end=end_dt.strftime('%Y-%m-%d'),
                interval='1d',
                progress=False,
                auto_adjust=True,
            )
            if df is not None and not df.empty:
                if isinstance(df.columns, pd.MultiIndex):
                    lv0 = df.columns.get_level_values(0).tolist()
                    lv1 = df.columns.get_level_values(1).tolist()
                    if ticker in lv1:
                        df.columns = [str(c).lower().strip() for c in lv0]
                    elif ticker in lv0:
                        df.columns = [str(c).lower().strip() for c in lv1]
                    else:
                        df.columns = ['_'.join(str(x) for x in c).lower().strip()
                                      for c in df.columns]
                else:
                    df.columns = [str(c).lower().strip() for c in df.columns]
                df = df.loc[:, df.columns.str.strip() != '']
                return df
        except Exception:
            pass
        time.sleep(0.5)
    return None


def calc_scores_for_date(
    df_ohlcv:     pd.DataFrame,
    close_col:    str,
    date_str:     str,
    fundamentals: Dict[str, Any],
    ticker:       str = '',
) -> Optional[Dict[str, Any]]:
    df_ohlcv.index = pd.to_datetime(df_ohlcv.index, errors='coerce')
    df_ohlcv       = df_ohlcv.sort_index()
    target_ts = pd.Timestamp(date_str)
    df_slice  = df_ohlcv[df_ohlcv.index <= target_ts]
    if len(df_slice) < 2:
        return None
    last_row = df_slice.iloc[-1]
    price    = float(last_row[close_col])
    if np.isnan(price) or price <= 0:
        return None
    df_slice = calc_moving_averages(df_slice, close_col)
    df_slice = calc_bollinger_bands(df_slice, close_col)
    df_slice = calc_rsi(df_slice, close_col)
    last     = df_slice.iloc[-1]
    required = ['25MA','50MA','75MA','RSI','BB_+1σ','BB_+2σ','BB_-1σ','BB_-2σ']
    if any(col not in df_slice.columns
           or np.isnan(float(last.get(col, np.nan))) for col in required):
        return None
    ma_25    = float(last['25MA'])
    ma_50    = float(last['50MA'])
    ma_75    = float(last['75MA'])
    rsi      = float(last['RSI'])
    bb_p1    = float(last['BB_+1σ'])
    bb_p2    = float(last['BB_+2σ'])
    bb_m1    = float(last['BB_-1σ'])
    bb_m2    = float(last['BB_-2σ'])
    slope_25 = calc_slope(df_slice['25MA'])
    df_1y    = df_slice.iloc[-252:]
    high_col = next((c for c in df_slice.columns if c.lower().startswith('high')), None)
    low_col  = next((c for c in df_slice.columns if c.lower().startswith('low')),  None)
    high_52w = float(df_1y[high_col].max()) if high_col else float(df_1y[close_col].max())
    low_52w  = float(df_1y[low_col].min())  if low_col  else float(df_1y[close_col].min())
    eps = fundamentals.get('eps')
    bps = fundamentals.get('bps')
    per = (price / eps) if eps and eps != 0 else None
    pbr = (price / bps) if bps and bps != 0 else None
    divs_1y        = fundamentals.get('_dividends_1y', 0.0) or 0.0
    dividend_yield = (divs_1y / price * 100.0) if price > 0 and divs_1y > 0 else None
    industry = fundamentals.get('industry') or ''
    sector   = fundamentals.get('sector')   or ''
    is_us    = not (ticker.endswith('.T') or ticker.replace('.T','').isdigit())
    q_result = score_quality(
        roe=fundamentals.get('roe'),
        roa=fundamentals.get('roa'),
        equity_ratio=fundamentals.get('equity_ratio'),
        operating_margin=fundamentals.get('operating_margin'),
        de_ratio=fundamentals.get('de_ratio'),
        interest_coverage=fundamentals.get('interest_coverage'),
        industry=industry, sector=sector, is_us=is_us,
    )
    q_score, q1, q3 = q_result['q_score'], q_result['q1'], q_result['q3']
    v_result = score_valuation(
        per=per, pbr=pbr, dividend_yield=dividend_yield,
        ev_ebitda=fundamentals.get('ev_ebitda'), is_us=is_us,
    )
    v_score = v_result['v_score']
    v1, v2, v3 = v_result['v1'], v_result['v2'], v_result['v3']
    t_metrics = compute_t_metrics(
        price=price, ma_25=ma_25, ma_50=ma_50, ma_75=ma_75,
        rsi=rsi, bb_plus1=bb_p1, bb_plus2=bb_p2,
        bb_minus1=bb_m1, bb_minus2=bb_m2,
        slope_25=slope_25,
        low_52w=low_52w, high_52w=high_52w,
        per=per, pbr=pbr,
    )
    t_score   = t_metrics['t_score']
    qvt_score = round((q_score + v_score + t_score) / 3.0, 1)
    return {
        'close': price, 'ma_25': ma_25, 'ma_50': ma_50,
        'rsi':   rsi,   'high_52w': high_52w, 'low_52w': low_52w,
        'per': per, 'pbr': pbr,
        'q_score': q_score, 'q1': q1, 'q3': q3,
        'v_score': v_score, 'v1': v1, 'v2': v2, 'v3': v3,
        't_score': t_score, 'qvt_score': qvt_score,
        'timing_label': t_metrics.get('timing_label', ''),
        'signal_text':  t_metrics.get('signal_text',  ''),
    }


print('Utility functions defined.')
````

````python
# ===================================================
# Cell 6: Per-ticker snapshot function
#   Self-contained: fetch OHLCV + fundamentals,
#   compute scores, return one row dict.
#   Safe for parallel execution.
# ===================================================

def process_one_ticker(ticker: str, date_str: str,
                        sleep_sec: float = 0.5) -> Dict[str, Any]:
    row: Dict[str, Any] = {
        'date':            date_str,
        'quarter':         get_quarter_key(date_str),
        'ticker':          ticker,
        'company_name':    ticker,
        'close':           NA, 'ma_25': NA, 'ma_50': NA,
        'rsi':             NA, 'high_52w': NA, 'low_52w': NA,
        'per':             NA, 'pbr': NA,
        'q_score':         NA, 'q1': NA, 'q3': NA,
        'v_score':         NA, 'v1': NA, 'v2': NA, 'v3': NA,
        't_score':         NA, 'qvt_score': NA,
        'timing_label':    NA, 'signal_text': NA,
        'roe':             NA, 'roa': NA, 'equity_ratio': NA,
        'operating_margin': NA, 'de_ratio': NA,
        'interest_coverage': NA, 'ev_ebitda': NA,
        'industry':        NA, 'sector': NA,
        '_status':         'pending',
    }
    try:
        # 1) fundamentals
        fund = fetch_fund_for_date(ticker, date_str)
        row['company_name']      = fund.get('company_name') or ticker
        row['roe']               = _safe(fund.get('roe'))
        row['roa']               = _safe(fund.get('roa'))
        row['equity_ratio']      = _safe(fund.get('equity_ratio'))
        row['operating_margin']  = _safe(fund.get('operating_margin'))
        row['de_ratio']          = _safe(fund.get('de_ratio'))
        row['interest_coverage'] = _safe(fund.get('interest_coverage'))
        row['ev_ebitda']         = _safe(fund.get('ev_ebitda'))
        row['industry']          = fund.get('industry', NA) or NA
        row['sector']            = fund.get('sector',   NA) or NA

        # 2) OHLCV
        df_ohlcv = fetch_ohlcv_for_date(ticker, date_str)
        if sleep_sec > 0:
            time.sleep(sleep_sec)
        if df_ohlcv is None or df_ohlcv.empty:
            row['_status'] = 'no_ohlcv'
            return row
        close_col = get_close_col(df_ohlcv)
        if close_col is None:
            row['_status'] = 'no_close_col'
            return row

        # 3) score calculation
        scores = calc_scores_for_date(df_ohlcv, close_col, date_str, fund, ticker)
        if scores is None:
            row['_status'] = 'holiday_or_insufficient'
            return row

        row.update({
            'close':        _safe(scores['close'], 0),
            'ma_25':        _safe(scores['ma_25'], 0),
            'ma_50':        _safe(scores['ma_50'], 0),
            'rsi':          _safe(scores['rsi'], 1),
            'high_52w':     _safe(scores['high_52w'], 0),
            'low_52w':      _safe(scores['low_52w'], 0),
            'per':          _safe(scores['per'], 1),
            'pbr':          _safe(scores['pbr'], 2),
            'q_score':      _safe(scores['q_score'], 1),
            'q1':           _safe(scores['q1'], 1),
            'q3':           _safe(scores['q3'], 1),
            'v_score':      _safe(scores['v_score'], 1),
            'v1':           _safe(scores['v1'], 1),
            'v2':           _safe(scores['v2'], 1),
            'v3':           _safe(scores['v3'], 1),
            't_score':      _safe(scores['t_score'], 1),
            'qvt_score':    _safe(scores['qvt_score'], 1),
            'timing_label': scores.get('timing_label', NA) or NA,
            'signal_text':  scores.get('signal_text',  NA) or NA,
        })
        row['_status'] = 'ok'
    except Exception as e:
        row['_status'] = f'error: {e}'
    return row


print('process_one_ticker defined.')
````

````python
# ===================================================
# Cell 7: Main execution -- full TSE snapshot
# ===================================================

from IPython.display import display, HTML
from concurrent.futures import ThreadPoolExecutor, as_completed

OUTPUT_CSV = Path(OUTPUT_DIR) / f'snapshot_{TARGET_DATE.replace("-","")}.csv'
total      = len(TICKERS)
results    = []
status_counter = {'ok': 0, 'no_ohlcv': 0, 'holiday_or_insufficient': 0, 'error': 0}

print(f'Snapshot start: {TARGET_DATE}')
print(f'Tickers       : {total}')
print(f'Parallel      : {"ON (workers=" + str(MAX_WORKERS) + ")" if USE_PARALLEL else "OFF"}')
print()

t_start = time.time()

if USE_PARALLEL:
    futures_map = {}
    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        for ticker in TICKERS:
            f = executor.submit(process_one_ticker, ticker, TARGET_DATE, SLEEP_SEC)
            futures_map[f] = ticker
        done = 0
        for future in as_completed(futures_map):
            done += 1
            ticker = futures_map[future]
            try:
                row = future.result()
            except Exception as e:
                row = {'ticker': ticker, '_status': f'future_error: {e}'}
            results.append(row)
            st  = row.get('_status', 'unknown')
            key = st if st in status_counter else 'error'
            status_counter[key] += 1
            if done % 100 == 0 or done == total:
                elapsed = time.time() - t_start
                eta     = (elapsed / done) * (total - done)
                print(f'  {done:4d}/{total}  '
                      f'ok={status_counter["ok"]} '
                      f'err={status_counter["error"]} '
                      f'elapsed={elapsed/60:.1f}min '
                      f'eta={eta/60:.1f}min')
else:
    for i, ticker in enumerate(TICKERS, 1):
        row = process_one_ticker(ticker, TARGET_DATE, SLEEP_SEC)
        results.append(row)
        st  = row.get('_status', 'unknown')
        key = st if st in status_counter else 'error'
        status_counter[key] += 1
        if i % 100 == 0 or i == total:
            elapsed = time.time() - t_start
            eta     = (elapsed / i) * (total - i)
            print(f'  {i:4d}/{total} {ticker}  '
                  f'ok={status_counter["ok"]} '
                  f'err={status_counter["error"]} '
                  f'elapsed={elapsed/60:.1f}min '
                  f'eta={eta/60:.1f}min')

# Save
snapshot_df  = pd.DataFrame(results)
save_cols    = [c for c in snapshot_df.columns if not c.startswith('_')]
snapshot_df[save_cols].to_csv(OUTPUT_CSV, index=False, encoding='utf-8-sig')

elapsed_total = time.time() - t_start
print(f'\nDone! {elapsed_total/60:.1f} min')
print(f'  ok          : {status_counter["ok"]}')
print(f'  error       : {status_counter["error"]}')
print(f'  no_ohlcv    : {status_counter["no_ohlcv"]}')
print(f'  holiday     : {status_counter["holiday_or_insufficient"]}')
print(f'  saved to    : {OUTPUT_CSV}')
````

````python
# ===================================================
# Cell 8: Results summary
# ===================================================

from IPython.display import display

if 'snapshot_df' not in dir():
    snapshot_df = pd.read_csv(OUTPUT_CSV, encoding='utf-8-sig')

ok_df = snapshot_df[snapshot_df['close'] != '-'].copy()
for col in ['close','q_score','v_score','t_score','qvt_score','rsi','per','pbr']:
    ok_df[col] = pd.to_numeric(ok_df[col], errors='coerce')

print(f'Total rows     : {len(snapshot_df)}')
print(f'Scores computed: {len(ok_df)}')
print(f'No score       : {len(snapshot_df) - len(ok_df)}')
print()

# QVT top 30
top30 = ok_df.nlargest(30, 'qvt_score')[[
    'ticker','company_name','qvt_score','t_score','q_score','v_score',
    'close','rsi','timing_label','signal_text','sector'
]]
print('=== Top 30 by QVT score ===')
display(top30.reset_index(drop=True))
````

````python
# ===================================================
# Cell 9: Filtering & sector breakdown
# ===================================================

# Bargain candidates: QVT>=50, T>=70, Q>=40
bargain = ok_df[
    (ok_df['qvt_score'] >= 50) &
    (ok_df['t_score']   >= 70) &
    (ok_df['q_score']   >= 40)
].sort_values('qvt_score', ascending=False)

print(f'=== Bargain candidates (QVT>=50, T>=70, Q>=40): {len(bargain)} ===')
show_cols = ['ticker','company_name','qvt_score','t_score','q_score','v_score',
             'close','rsi','timing_label','sector']
display(bargain[show_cols].head(30).reset_index(drop=True))

# Sector average
print('\n=== Sector average QVT ===')
sector_avg = (
    ok_df.groupby('sector')['qvt_score']
    .agg(['mean','count'])
    .rename(columns={'mean':'avg_qvt','count':'n'})
    .sort_values('avg_qvt', ascending=False)
    .head(20)
)
display(sector_avg.round(1))

# Score distribution
print('\n=== QVT distribution ===')
bins   = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = [f'{a}-{b}' for a, b in zip(bins[:-1], bins[1:])]
ok_df['qvt_bin'] = pd.cut(ok_df['qvt_score'], bins=bins, labels=labels, right=False)
print(ok_df['qvt_bin'].value_counts().sort_index().to_string())
````

````python
# ===================================================
# Cell 10: Plotly visualization
# ===================================================

try:
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    PLOTLY_OK = True
except ImportError:
    PLOTLY_OK = False
    print('plotly not installed. Run: pip install plotly')

THEME = dict(
    paper_bgcolor='#0a0e1a',
    plot_bgcolor='#111827',
    font=dict(color='#94a3b8', family='sans-serif'),
)

if PLOTLY_OK and 'ok_df' in dir() and len(ok_df) > 0:
    plot_df = ok_df.dropna(subset=['qvt_score','t_score','q_score'])

    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=(
            f'QVT score distribution ({TARGET_DATE}, n={len(plot_df)})',
            'T-score vs Q-score (color=QVT)',
        ),
    )
    fig.add_trace(go.Histogram(
        x=plot_df['qvt_score'], nbinsx=40,
        marker_color='#3b82f6', name='QVT dist',
    ), row=1, col=1)
    fig.add_trace(go.Scatter(
        x=plot_df['t_score'], y=plot_df['q_score'],
        mode='markers',
        marker=dict(
            color=plot_df['qvt_score'], colorscale='Viridis',
            size=5, opacity=0.6, showscale=True,
            colorbar=dict(title='QVT', x=1.05),
        ),
        text=plot_df['ticker'],
        hovertemplate='%{text}<br>T=%{x:.1f} Q=%{y:.1f}',
        name='T vs Q',
    ), row=1, col=2)
    fig.update_layout(
        **THEME, height=420,
        title=dict(text=f'TSE Snapshot {TARGET_DATE}',
                   font=dict(size=16, color='#f1f5f9')),
        showlegend=False,
        margin=dict(l=0, r=60, t=60, b=0),
    )
    fig.update_xaxes(gridcolor='#1e293b')
    fig.update_yaxes(gridcolor='#1e293b')
    fig.show()
````
